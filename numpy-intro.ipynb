## 1. Array Attributes

Let's say `arr` is a numpy array. These are some key attributes:

* **Shape**: `arr.shape` denotes the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. 


* **Size**: `arr.size` denotes the total number of elements of the array.


* **Data-type**: `arr.dtype` denotes the type of the elements in the array. Some examples are `int32`, `int64`, `float32`, `float64`, `bool` and so on.


* **Number of Dimensions**: `arr.ndim` denotes the number of dimensions in the array.


In other words, we have 
```
arr.size == np.prod(arr.shape)
arr.ndim == len(arr.shape)
```

Also, see the next cell.

````python
a = np.array(3)
b = np.array([3])
c = np.array([[3]])
d = np.array([[[3]]])

print(f'a.shape ==> {a.shape}')
print(f'b.shape ==> {b.shape}')
print(f'c.shape ==> {c.shape}')
print(f'd.shape ==> {d.shape}')
````

**Answer**: Indeed all of them are different, even though they all have identical underlying data. Even if the shapes only differ in some dummy dimensions of one, they're still considered to be different. 

These dummy dimensions of one do impact the alignment of dimensions when broadcasting; different shapes can lead to different operation/function behaviors.

## Broadcasting

It's just "virtual stretching" across dummy dimensions of size one:

<img src="08_bcast.png" alt="Drawing" style="width: 50%;"/>

Credits: https://numpy.org/doc/stable/user/basics.broadcasting.html

````python
a = np.array([1, 2, 3])
assert a.shape == (3,)

b = np.array([2])
assert b.shape == (1,)

c = a * b
assert c.shape == (3,)

c
````

<img src="09_bcast.png" alt="Drawing" style="width: 50%;"/>

````python
a = np.array([[ 0,  0,  0],
              [10, 10, 10],
              [20, 20, 20],
              [30, 30, 30]])
assert a.shape == (4, 3)

b = np.array([[1, 2, 3]])
assert b.shape == (1, 3)

c = a + b
assert c.shape == (4, 3)

c
````

Two arrays can have no identical dimensions, yet still be broadcastable!

<img src="10_bcast.png" alt="Drawing" style="width: 50%;"/>

<font color=red>Challenge</font>: Replace each pixel with the average of its neighbors. Do not use convolutions.

````python
import numpy as np
import matplotlib.pyplot as plt

# Assuming red_channel, green_channel, and blue_channel are 2D numpy arrays of the same shape

# Stack the RGB channels into a 3D array
rgb_image = np.stack((red_channel, green_channel, blue_channel), axis=-1)

# Define the weights for the luminosity method
weights = np.array([0.2126, 0.7152, 0.0722])

# Convert the RGB image to grayscale using Einstein summation
gray4 = np.einsum('ijk,k->ij', rgb_image, weights)

# Pad the grayscale image to handle border pixels
padded_gray = np.pad(gray4, pad_width=1, mode='edge')

# Create the output array for the result
output_image = np.zeros_like(gray4)

# Replace each pixel with the average of its neighbors
for i in range(1, padded_gray.shape[0] - 1):
    for j in range(1, padded_gray.shape[1] - 1):
        # Extract the 3x3 neighborhood
        neighbors = padded_gray[i-1:i+2, j-1:j+2]
        
        # Compute the average of the neighbors
        output_image[i-1, j-1] = np.mean(neighbors)

# Display the original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray4, cmap='gray')
plt.title('Grayscale Image')

# Display the processed image with neighbor averaging
plt.subplot(1, 2, 2)
plt.imshow(output_image, cmap='gray')
plt.title('Average of Neighbors')

plt.show()
````

* In each grid, find the pixel with maximum luminance.

````python
import numpy as np
import matplotlib.pyplot as plt

# Assuming red_channel, green_channel, and blue_channel are 2D numpy arrays of the same shape

# Stack the RGB channels into a 3D array
rgb_image = np.stack((red_channel, green_channel, blue_channel), axis=-1)

# Define the weights for the luminosity method
weights = np.array([0.2126, 0.7152, 0.0722])

# Convert the RGB image to grayscale using Einstein summation
gray4 = np.einsum('ijk,k->ij', rgb_image, weights)

# Pad the grayscale image to handle border pixels
padded_gray = np.pad(gray4, pad_width=1, mode='edge')

# Replace each pixel with the average of its neighbors using array operations
output_image = (
    padded_gray[:-2, :-2] + padded_gray[:-2, 1:-1] + padded_gray[:-2, 2:] +
    padded_gray[1:-1, :-2] + padded_gray[1:-1, 1:-1] + padded_gray[1:-1, 2:] +
    padded_gray[2:, :-2] + padded_gray[2:, 1:-1] + padded_gray[2:, 2:]
) / 9

# Display the original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray4, cmap='gray')
plt.title('Grayscale Image')

# Display the processed image with neighbor averaging
plt.subplot(1, 2, 2)
plt.imshow(output_image, cmap='gray')
plt.title('Average of Neighbors')

plt.show()
````

<font color=red>Challenge</font>: Let's say you have an $N\times D$ data matrix and a bunch of labels $y$. You have $C$ classes in your data.
    
Find the data mean within each class. You can only use numpy operations.

````python
N = 100
D = 8
C = 10

np.random.seed(12345)
X = np.random.randn(N, D)
assert X.shape == (N, D)
Y = np.random.randint(0, C, N)
assert Y.shape == (N,)
````

````python
import numpy as np

# Given data dimensions
N = 100  # Number of data points
D = 8    # Dimensionality of each data point
C = 10   # Number of classes

# Initialize the random seed and generate the data matrix and labels
np.random.seed(12345)
X = np.random.randn(N, D)  # Data matrix of shape (N, D)
Y = np.random.randint(0, C, N)  # Labels array of shape (N,)

# Step 1: One-hot encode the labels Y to create an indicator matrix
Y_one_hot = np.eye(C)[Y]  # Shape (N, C)

# Step 2: Compute the sum of data points in each class using matrix multiplication
class_sums = np.dot(Y_one_hot.T, X)  # Shape (C, D)

# Step 3: Count the number of data points in each class
class_counts = np.sum(Y_one_hot, axis=0).reshape(-1, 1)  # Shape (C, 1)

# Step 4: Calculate the mean by dividing the sum by the count
class_means = class_sums / class_counts  # Shape (C, D)

# Output the class means
print("Class Means:\n", class_means)
````
